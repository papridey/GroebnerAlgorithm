```python
import sympy

R.<x,y,z,s,w,l> = PolynomialRing(QQ,6,order='degrevlex')




def symbolicpreprocessing(L,G):
    F ={i[0]*i[1] for i in L}
    Done = {j.lm() for j in F}
    
    Term = lambda f: {item for sublist in f for item in sublist.monomials()}
    Terms = Term(F)
    comp = Terms - Done
    while Terms != Done:
        for m in comp:
            break
           
        Done = Done|{m}
        for f in G:
            if m.reduce([f.lm()]) == 0:
                deg = f.lm().degrees()
                M_deg = m.lm().degrees()
                M = x^(M_deg[0] - deg[0]) * y^(M_deg[1] - deg[1]) * z^(M_deg[2] - deg[2]) * s^(M_deg[3] - deg[3]) * w^(M_deg[4] - deg[4]) * l^(M_deg[5] - deg[5])
                Terms = Terms|Term([M * f])
                F = F|{M * f}
                break
        comp = Terms - Done
    
    
    H = [n for n in F]
    return H
    
def replace(F,L):
    M = []
    k = 0
    for n in L:
        k = 0
        for j in F:
            if j[0] == n:
                k = j[1]
                break
        M.append(k)
    return M
    
    

    
def rowechelon(F):
    G=[]
    for a in F:
        G.append([[b,a.coefficients()[c]] for c,b in enumerate(a.exponents())])
    exponentiate = lambda f: x^f[0] * y^f[1] * z^f[2] * s^f[3] * w^f[4] * l^f[5]
    powers = lambda f: {item for sublist in f for item in sublist.exponents()}
    L = [b for b in powers(F)]
    L = sorted(L, key = exponentiate, reverse = true)
    M = sympy.Matrix([replace(a,L) for a in G])
    print('size of matrix =',len(M))
    H = []
    M = M.rref()
    Leading_terms = [o[0][0] for o in G]
    for c,v in enumerate(M[1]):
        if L[v] not in Leading_terms:
            t = 0
            for b,e in enumerate(M[0].row(c)):
                if e != 0:
                    #casting from sympy object to sagemath polynomial
                    r = e + x
                    e = r - x
                    
                    t = t + (e * exponentiate(L[b]))
            H.append(t)
            
    return H

def reduction(L,G):
    F = symbolicpreprocessing(L,G)
    H = rowechelon(F)
    return H

def Pair(f,g):
    exponentiate = lambda f: x^f[0]*y^f[1]*z^f[2] * s^f[3] * w^f[4] * l^f[5]
    subtract = lambda f,g: [f[0] - g[0], f[1] - g[1], f[2] - g[2], f[3] - g[3], f[4] - g[4], f[5] - g[5]] 
    h = lcm(f.lm(),g.lm())
    f_1 = exponentiate(subtract(h.exponents()[0],f.exponents()[0]))
    g_1 = exponentiate(subtract(h.exponents()[0],g.exponents()[0]))
    return [h,f_1,f,g_1,g]

def cPair(F):
    return([Pair(a, b) for c, a in enumerate(F) for b in F[c + 1:]])


def sel(P):
    exponentiate = lambda f: x^f[0]*y^f[1]*z^f[2] *s^f[3] * w^f[4] * l^f[5]
    d = [exponentiate(a[0].degrees()) for a in P]
    minimum = min(d).degrees()
    P_0 = [b for b in P if b[0].degrees() == minimum ]
    return P_0


def F4(F):
    G = F
    F_0 = F
    d = 0
    P = cPair(G)
    Left = lambda F: [[q[1],q[2]] for q in F]
    Right = lambda F: [[q[3],q[4]] for q in F]
    while len(P) != 0:
        P_d = sel(P)
        P = [a for a in P if a not in P_d]
        L_d = Left(P_d) + Right(P_d)
        F_d = reduction(L_d,G)
        
        for h in F_d:
            for g in G:
                if Pair(h,g) not in P:
                    P.append(Pair(h,g))
            if h not in G:
                G.append(h)
       
          
    return G



a = [x + y + z +s + w + l, x*y*z*s*w*l - 1, x*y + y*z + z*s + s*w + w*l + l*x,
    x*y*z + y*z*s + z*s*w + s*w*l + w*l*x + l*x*y,
    x*y*z*s + y*z*s*w + z*s*w*l + s*w*l*x + w*l*x*y + l*x*y*z,
    x*y*z*s*w + y*z*s*w*l + z*s*w*l*x + s*w*l*x*y + w*l*x*y*z + l*x*y*z*s]
print(F4(a))
```
